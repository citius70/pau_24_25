## 3. Visualización de los datos

Generamos visualizaciones para explorar la distribución de los datos. Los gráficos seleccionados son histogramas y diagramas de caja (box plot).


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select numerical columns for visualization
numerical_cols = df.select_dtypes(include=['number'])
columns_to_exclude = ['Cod. Lingua Estranxeira', 'Cod. Materia Modalidade'] # Corrected column name
numerical_cols_viz = numerical_cols.drop(columns=columns_to_exclude)

# Calculate the mean excluding the last two columns
mean_excluding_last_two = numerical_cols_viz.iloc[:, :-2].mean().mean()

# Calculate the overall mean of all numerical columns being visualized
overall_mean_numerical_viz = numerical_cols_viz.mean().mean()


# Create histograms
numerical_cols_viz.hist(bins=10, figsize=(15, 10))
plt.suptitle('Histogramas de Columnas Numéricas', y=0.92, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.grid(axis='y', linestyle='--') # Add horizontal grid lines
plt.xlim(0, 10) # Set x-axis limits from 0 to 10
plt.show()

# Create box plots
plt.figure(figsize=(10, 8))
sns.boxplot(data=numerical_cols_viz)
plt.title('Box Plots de Columnas Numéricas')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(0, 11, 1)) # Set y-axis ticks from 0 to 10 with step 1
plt.ylim(0, 10) # Set y-axis limits from 0 to 10
plt.xlabel('MATERIAS') # Set x-axis label
plt.ylabel('PUNTUACIONES: Mínimo, 2º cuartil, media, 3º cuartil, máximo') # Set y-axis label

# Add horizontal line for the mean excluding the last two columns
plt.axhline(mean_excluding_last_two, color='red', linestyle='dashed', linewidth=1, label=f'Media (excluyendo últimas 2 columnas) ({mean_excluding_last_two:.2f})')

# Add horizontal line for the overall mean of all numerical columns being visualized
plt.axhline(overall_mean_numerical_viz, color='blue', linestyle='dotted', linewidth=1, label=f'Media General ({overall_mean_numerical_viz:.2f})')


plt.legend() # Show the legend for the horizontal lines
plt.tight_layout()
plt.show()
```
