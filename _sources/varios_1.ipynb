## 1. Heatmap de correlación

Se crea un heatmap para visualizar la matriz de correlación entre las variables numéricas. Esto nos permitirá identificar qué variables están más relacionadas entre sí.

```python
import pandas as pd

df = pd.read_excel("proba_xeral.xlsx")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seleccionar columnas numéricas
num_cols = df.select_dtypes(include=['number'])
cols_excluir = ['Cod. Lingua Estranxeira', 'Cod. Materia Modalidade', 'Cualificación parte xeral', 'Cualificación expediente bacharelato']
num_cols_viz = num_cols.drop(columns=cols_excluir)

# Eliminar columnas con todos los valores NaN
num_cols_viz = num_cols_viz.dropna(axis=1, how='all')

# Calcular la matriz de correlación
correlation_matrix = num_cols_viz.corr()

# Crear el heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='rocket_r', fmt=".2f", linewidths=0,vmin=-1, vmax=1)
plt.title('Heatmap de correlación de las calificaciones')
plt.show()
```

### Análisis de la Matriz de Correlación

La matriz muestra las correlaciones entre las calificaciones de diferentes asignaturas, excluyendo "Cualificación parte xeral" y "Expediente bacharelato". Los valores de correlación oscilan entre -0.36 y 1.00, donde 1 indica una correlación perfecta positiva, -1 una correlación perfecta negativa, y 0 ausencia de correlación.

#### Correlaciones destacadas:
1. **Língua castelá e literatura e Historia de España**:  
   - **Correlación: 0.56**  
   - Es la correlación más alta, lo que sugiere que los estudiantes que obtienen buenas calificaciones en una de estas asignaturas tienden a desempeñarse bien en la otra. Esto podría deberse a habilidades compartidas, como comprensión lectora o análisis crítico.

2. **Língua galega e literatura e Historia da Filosofía**:  
   - **Correlación: 0.42**  
   - Indica una relación moderada, posiblemente vinculada a competencias lingüísticas y de razonamiento abstracto.

3. **Historia da Filosofía e Língua Estranxeira**:  
   - **Correlación: -0.36**  
   - Es la correlación negativa más fuerte, lo que sugiere que los estudiantes que sobresalen en una de estas asignaturas tienden a tener dificultades en la otra. Esto podría reflejar diferencias en los perfiles de aprendizaje o intereses académicos.

4. **Língua Estranxeira e Materia Modalidade**:  
   - **Correlación: 0.29**  
   - Aunque positiva, es relativamente débil, lo que indica que el desempeño en una asignatura no necesariamente predice el desempeño en la otra.

5. **Língua castelá e literatura e Língua Estranxeira**:  
   - **Correlación: 0.04**  
   - Prácticamente nula, lo que sugiere que no hay una relación discernible entre el desempeño en estas dos asignaturas.

### Conclusiones:
- **Asignaturas con mayor relación**: Las correlaciones más altas se observan entre asignaturas de humanidades (Língua castelá e literatura, Historia de España, Língua galega e literatura), lo que podría indicar que comparten habilidades fundamentales o que atraen a estudiantes con perfiles similares.
  
- **Asignaturas con relación inversa**: La correlación negativa entre Historia da Filosofía y Língua Estranxeira sugiere que los estudiantes pueden tener preferencias o aptitudes divergentes para estas materias.

- **Asignaturas independientes**: Língua Estranxeira muestra correlaciones bajas o negativas con varias asignaturas, lo que podría indicar que requiere un conjunto de habilidades distintivo.

### Recomendaciones:
- **Enfoque pedagógico**: Los profesores podrían considerar estrategias interdisciplinarias para asignaturas con correlaciones altas, como integrar contenidos de Historia de España y Língua castelá e literatura.
  
- **Orientación estudiantil**: Los estudiantes que tienen dificultades en asignaturas con correlaciones negativas podrían beneficiarse de tutorías personalizadas que aborden sus estilos de aprendizaje específicos.

- **Investigación adicional**: Sería útil explorar factores externos (como horas de estudio o métodos de enseñanza) que puedan influir en estas correlaciones.

En resumen, la matriz revela patrones interesantes en el desempeño académico, destacando la interconexión entre algunas asignaturas y la independencia de otras.
