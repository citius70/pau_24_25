## 2. Visualización por materia

Generar histogramas y box plots de la columna 'Materia Modalidade' para cada materia de modalidad para visualizar sus distribuciones.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

# Get unique modalities and create a color mapping using 'Set3'
modalities = modalidad_df['Nom. Materia modalidade'].unique()
colors = sns.color_palette('Set3', len(modalities))
color_map = dict(zip(modalities, colors))

# Histograms
num_modalidades = len(modalities)
cols = 3
rows = math.ceil(num_modalidades / cols)

plt.figure(figsize=(cols * 5, rows * 4))

for i, modality in enumerate(modalities):
    plt.subplot(rows, cols, i + 1)
    group = grouped_modalidad_df.get_group(modality)
    plt.hist(group, bins=10, color=color_map[modality]) # Use color map
    plt.title(f'Distribución de Cualificacións para {modality}')
    plt.xlabel('Cualificación')
    plt.ylabel('Frecuencia')
    plt.xlim(0, 10)
    plt.ylim(0, 4) # Set fixed y-axis limit
    plt.yticks(np.arange(0, 5, 1)) # Set fixed y-ticks

    plt.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()

# Box plot with individual points and means
plt.figure(figsize=(10, 6))
sns.boxplot(data=modalidad_df, x='Nom. Materia modalidade', y='Materia Modalidade', hue='Nom. Materia modalidade', palette=color_map, legend=False) # Use color map
sns.swarmplot(data=modalidad_df, x='Nom. Materia modalidade', y='Materia Modalidade', color='black', size=5, ax=plt.gca()) # Add individual points

# Add mean points for each modality
modalities_mean = grouped_modalidad_df.mean()
plt.scatter(x=range(len(modalities)), y=modalities_mean.values, color='white', edgecolors='black', s=100, zorder=3) # Add white points with black edge

plt.title('Box Plot de Cualificacións por Materia de Modalidade con Puntos Individuais e Medias')
plt.xlabel('Materia de Modalidade')
plt.ylabel('Cualificación')
plt.ylim(0, 10)
plt.xticks(rotation=45, ha='right')



plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

# Histograms
num_modalidades = len(grouped_modalidad_df)
cols = 3
rows = math.ceil(num_modalidades / cols)

plt.figure(figsize=(cols * 5, rows * 4))
for i, (modality, group) in enumerate(grouped_modalidad_df):
    plt.subplot(rows, cols, i + 1)
    counts, bins, patches = plt.hist(group, bins=10)
    plt.title(f'Distribución de Cualificacións para {modality}')
    plt.xlabel('Cualificación')
    plt.ylabel('Frecuencia')
    plt.xlim(0, 10)
    plt.yticks(np.arange(0, int(max(counts)) + 1, 1))
    plt.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=modalidad_df, x='Nom. Materia modalidade', y='Materia Modalidade', hue='Nom. Materia modalidade', palette='viridis', legend=False)
plt.title('Box Plot de Cualificacións por Materia de Modalidade')
plt.xlabel('Materia de Modalidade')
plt.ylabel('Cualificación')
plt.ylim(0, 10) # Adjusted y-limit for better visualization
plt.xticks(rotation=45, ha='right')

# Add overall mean and median lines
overall_mean = modalidad_df['Materia Modalidade'].mean()
plt.axhline(overall_mean, color='red', linestyle='dashed', linewidth=1, label=f'Media Materias Modalidade ({overall_mean:.2f})')

overall_median = modalidad_df['Materia Modalidade'].median()
plt.axhline(overall_median, color='green', linestyle='dashed', linewidth=1, label=f'Mediana Materias Modalidade ({overall_median:.2f})')

plt.legend()
plt.tight_layout()
plt.show()
```
