### **Análisis del Histograma:**
1. **Distribución general**:  
   - **Pico principal**: La frecuencia más alta se concentra en las calificaciones **6-8**, lo que indica que la mayoría de los estudiantes obtienen notas medias-altas.  
   - **Cola izquierda (notas bajas)**: Hay frecuencias bajas pero significativas en el rango **0-4**, sugiriendo un grupo minoritario con dificultades.  
   - **Cola derecha (notas altas)**: Las notas **9-10** tienen menos frecuencia que el pico central, pero aún destacan.  

2. **Moda (valor más frecuente)**:  
   - Probablemente alrededor de **6**, dado que es el intervalo con mayor frecuencia.  

3. **Asimetría**:  
   - La distribución está **levemente sesgada a la izquierda** (más notas altas que bajas), pero con una cola izquierda visible (posibles casos problemáticos).  

4. **Brechas o patrones inesperados**:  
   - Si hay intervalos con **frecuencia cercana a 0** (ej. calificación 5), podría indicar:  
     - Polarización (estudiantes que o bien aprueban bien o suspenden).  
     - Efecto de redondeo en las notas (ej. profesores que evitan notas intermedias).  

---

### **Conclusiones**:  
- **Rendimiento mayoritariamente satisfactorio**: La mayoría de las calificaciones se ubican en el rango de aprobado/notable (6-8).  
- **Grupo vulnerable**: Las notas bajas (0-4) representan un subconjunto que requiere atención (ej. refuerzo educativo o revisión de metodologías).  
- **Excelencia limitada**: Pocos estudiantes alcanzan notas máximas (9-10), lo que podría sugerir que las materias opcionales son desafiantes o los criterios de evaluación son estrictos.

### **Análisis del Boxplot:**
1. **Rango de calificaciones**:  
   - Las calificaciones abarcan desde **0 hasta 10**, como es habitual en sistemas educativos.  
   - La distribución parece estar **sesgada hacia calificaciones medias-altas** (ej. concentración entre 5 y 9), aunque hay outliers en el extremo inferior (ej. 0 o 1).

2. **Mediana (línea central en la caja)**:  
   - Próxima al **6**, lo que sugiere que la mayoría de los estudiantes obtienen calificaciones aprobatorias aceptables.

3. **Disperión (rango intercuartílico - IQR)**:  
   - La caja (que representa el 50% central de los datos) abarca desde ~4 hasta ~7, indicando **variabilidad bastante moderada** en el desempeño.

4. **Outliers (puntos individuales)**:  
   - Hay calificaciones muy bajas (ej. cercanas a 0) que se alejan del resto, lo que podría indicar:  
     - Casos atípicos (estudiantes con dificultades).  
     - Errores en la evaluación.  
     - Dificultad extrema en alguna materia específica (si el gráfico agrupa múltiples asignaturas).

5. **Asimetría**:  
   - La cola inferior (valores bajos) es más larga que la superior, sugiriendo un **sesgo negativo** (más estudiantes en el rango alto, pero con algunos casos extremadamente bajos).

---

### **Conclusiones**:  
- **Rendimiento general aceptable**: La mayoría de las calificaciones se concentran en la zona media-alta.  
- **Posibles problemas específicos**: Los outliers bajos podrían señalar necesidades de apoyo educativo o revisión de métodos de enseñanza en ciertos casos.  
- **Consistencia**: Si el gráfico agrupa múltiples materias, la dispersión sugiere que el desempeño varía significativamente entre ellas (ej. materias con notas más altas vs. más bajas).

## Generar histogramas por materia

### Subtask:
Iterar sobre cada materia única y crear un histograma de las calificaciones correspondientes.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


materias_unicas = counts_df['Materia'].unique().tolist()


colors = sns.color_palette('Set1', len(materias_unicas))


n_materias = len(materias_unicas)
n_cols = 2
n_rows = (n_materias + n_cols - 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, n_rows * 4))
axes = axes.flatten()


consistent_bins = np.arange(0, 11, 1)

for i, materia in enumerate(materias_unicas):
    df_materia = df_cleaned[df_cleaned['Materia'] == materia].copy()
    if not df_materia.empty:
        ax = axes[i]
        sns.histplot(df_materia['Calificación'], bins=consistent_bins, kde=True, color=colors[i], ax=ax)
        ax.set_title(f'Distribución de Calificaciones para {materia}')
        ax.set_xlabel('Calificación')
        ax.set_ylabel('Frecuencia')
        ax.set_xlim(0, 10)
        ax.set_xticks(np.arange(0, 11, 1))
        ax.set_ylim(0, 5)
        ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))


        y_min, y_max = ax.get_ylim()
        y_ticks = ax.get_yticks()

        for y_value in y_ticks:
            if y_value >= y_min and y_value <= y_max:
                 ax.axhline(y=y_value, color='grey', linestyle='--', linewidth=0.6, alpha=0.5)


for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
