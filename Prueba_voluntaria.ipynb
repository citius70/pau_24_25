```python
import pandas as pd

# Cargar los datos desde el archivo Excel
try:
    df = pd.read_excel('proba_voluntaria.xlsx')
    print("Archivo cargado exitosamente.")
    # Mostrar las primeras filas y la información del DataFrame para verificar la carga
    display(df)
   
except FileNotFoundError:
    print("Error: Asegúrate de que el archivo 'Proba_voluntaria.xlsx' se encuentra en la ubicación correcta.")
    df = None # Asignar None a df si el archivo no se encuentra
except Exception as e:
    print(f"Ocurrió un error al cargar el archivo: {e}")
    df = None # Asignar None a df si hay otro error
```

```python
# Convertir la columna 'Cualif. Mat Opt. 1' a numérica, manejando errores
df['Cualif. Mat Opt. 1'] = pd.to_numeric(df['Cualif. Mat Opt. 1'], errors='coerce')

# Realizar el análisis estadístico descriptivo de la columna 'Cualif. Mat Opt. 1'
descriptive_stats = df['Cualif. Mat Opt. 1'].describe()

# Mostrar los resultados
print("Análisis Estadístico Descriptivo para 'Cualif. Mat Opt. 1':")
display(descriptive_stats)
```

```python
# Convertir la columna 'Cualif. Mat Opt. 2' a numérica, manejando errores
df['Cualif. Mat Opt. 2'] = pd.to_numeric(df['Cualif. Mat Opt. 2'], errors='coerce')

# Agrupar por 'Mat. Opt. 2' y aplicar describe() a 'Cualif. Mat Opt. 2'
descriptive_stats_by_subject_opt2 = df.groupby('Mat. Opt. 2')['Cualif. Mat Opt. 2'].describe()

# Mostrar los resultados
print("Análisis Estadístico Descriptivo de 'Cualif. Mat Opt. 2' por Materia:")
display(descriptive_stats_by_subject_opt2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker

# Ensure unique_subjects_opt2 is defined, in case the user skipped the previous cell
if 'unique_subjects_opt2' not in locals():
    unique_subjects_opt2 = df['Mat. Opt. 2'].unique().tolist()

# Calculate the number of rows needed (2 histograms per row)
n_subjects_opt2 = len(unique_subjects_opt2)
n_cols = 2
n_rows = (n_subjects_opt2 + n_cols - 1) // n_cols

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, n_rows * 5)) # Adjust figsize as needed

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Get the color palette
palette = sns.color_palette('husl', n_subjects_opt2)

# Create a mapping from subject to color using the same order as unique_subjects_opt2
color_map_opt2 = dict(zip(unique_subjects_opt2, palette))


for i, subject in enumerate(unique_subjects_opt2):
    # Filter the DataFrame for the current subject
    df_subject_opt2 = df[df['Mat. Opt. 2'] == subject].dropna(subset=['Cualif. Mat Opt. 2'])

    # Get the color for the current subject from the color_map_opt2
    subject_color_opt2 = color_map_opt2.get(subject)

    # Create a histogram for the filtered data on the current axis with the corresponding color
    sns.histplot(df_subject_opt2['Cualif. Mat Opt. 2'], kde=True, bins=10, ax=axes[i], color=subject_color_opt2)
    axes[i].set_title(f'Distribución de Cualificaciones en {subject}')
    axes[i].set_xlabel('Cualificación')
    axes[i].set_ylabel('Frecuencia')
    axes[i].grid(axis='y', alpha=0.75)
    axes[i].set_xlim(0, 10) # Set the x-axis limit from 0 to 10
    axes[i].yaxis.set_major_locator(mticker.MaxNLocator(integer=True)) # Set y-axis ticks to integers


# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

# Análisis Estadístico de Calificaciones en Mat. Opt. 2 por Materia

## Resumen estadístico general

El análisis de la columna 'Cualif. Mat Opt. 2' revela calificaciones para 10 materias diferentes. Al igual que con la primera opción, el tamaño de la muestra varía considerablemente entre las materias, desde 1 hasta 13 estudiantes.

### Materias con mejor desempeño (media más alta):

2. **Tecnoloxía e Enxeñaría II**: 8.42 (n=3)

2. **Empresa e D.M. Negocio**: 2.65 (rango 5.75-9.5)
3. **Química**: 2.61 (rango 0.25-8.5)

### Materias con peor desempeño (media más baja):

2. **Matemáticas Ap. ás CCSS**: 3.04 (n=6)
3. **Deseño**: 5.00 (n=1)

## Hallazgos clave

1. **Problemas de tamaño muestral persistentes**:
   - La mayoría de las materias (7 de 10) tienen n≤3.
   - 2 materias tienen n=1 (sin desviación estándar calculable).
2. **Variabilidad en calificaciones**:
   - Algunas materias como Bioloxía, Empresa e D.M. Negocio y Química muestran una dispersión considerable en las calificaciones.
3. **Materias con alto rendimiento (muestras pequeñas)**:
   - Coro e Técnica Vocal II y Tecnoloxía e Enxeñaría II tienen medias altas, pero basadas en muestras muy pequeñas, lo que limita la generalización.
4. **Materias con bajo rendimiento (muestras pequeñas a moderadas)**:
   - Francés y Matemáticas Ap. ás CCSS muestran medias bajas. Matemáticas Ap. ás CCSS tiene un tamaño de muestra más razonable (n=6) para este hallazgo.

## Limitaciones

1. La interpretación de los resultados está fuertemente limitada por los pequeños tamaños muestrales en la mayoría de las materias.
2. No es apropiado realizar comparaciones estadísticas rigurosas entre materias con tamaños muestrales tan dispares.

## Recomendaciones

1. **Incrementar el tamaño muestral** para obtener resultados más fiables.
2. **Considerar la agrupación de materias** si se desea realizar análisis comparativos más robustos.
3. **Continuar monitorizando** las calificaciones en futuras convocatorias para identificar tendencias.
