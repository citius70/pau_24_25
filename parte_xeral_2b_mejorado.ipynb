# #  1. Identificar las materias de modalidad

Filtramos el DataFrame para incluir solo las columnas relevantes para las materias de modalidad, las cuales son: ´Materia de modalidade´ y ´Cualif. Materia de Modalidade´

```python
import pandas as pd

df = pd.read_excel("proba_xeral.xlsx")

modalidad_df = df[['Apelidos','Nome', 'Nom. Materia modalidade', 'Materia Modalidade']].copy()

display(modalidad_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


modalidad_counts = modalidad_df['Nom. Materia modalidade'].value_counts()

plt.figure(figsize=(8, 5))
ax = sns.barplot(x=modalidad_counts.index, y=modalidad_counts.values, hue=modalidad_counts.index, palette='Set3')
plt.title('Número de Estudiantes con Calificación por Materia de Modalidad')

plt.ylabel('Nº de Estudiantes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()


for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
modalidad_df.groupby('Nom. Materia modalidade')['Materia Modalidade'].describe().transpose()
```

# #  2. Agrupar por materia de modalidad


Agrupar los datos por la columna 'Materia modalidade' para preparar el análisis estadístico y la visualización por materia.

```python
modalidad_stats = modalidad_df['Materia Modalidade'].describe().to_frame().T
display(modalidad_stats)


plt.figure(figsize=(4, 4))
sns.boxplot(y=modalidad_df['Materia Modalidade'], color='skyblue')

plt.title('Distribución de Materia Modalidade')
plt.ylabel('Puntuaciones')  


plt.tight_layout()
plt.show()    


sns.set(style="whitegrid")

# Crear histogramaplt.figure(figsize=(5, 4))
sns.histplot(modalidad_df['Materia Modalidade'], bins=10, kde=True, color='skyblue')


plt.title('Histograma de Calificaciones - Materia Modalidade', fontsize=14)
plt.xlabel('Calificación', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)

plt.tight_layout()
plt.show()
```

# #  2. Visualización por materia

Generar histogramas y box plots de la columna 'Materia Modalidade' para cada materia de modalidad para visualizar sus distribuciones.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math


modalities = sorted(modalidad_df['Nom. Materia modalidade'].unique())
colors = sns.color_palette('Set3', len(modalities))
color_map = dict(zip(modalities, colors))


grouped_modalidad_df = modalidad_df.groupby('Nom. Materia modalidade')['Materia Modalidade']


num_modalidades = len(modalities)
cols = 3
rows = math.ceil(num_modalidades / cols)

plt.figure(figsize=(cols * 5, rows * 4))

for i, modality in enumerate(modalities):
    plt.subplot(rows, cols, i + 1)
    group = grouped_modalidad_df.get_group(modality)
    plt.hist(group, bins=10, color=color_map[modality]) 
    plt.title(f'Distribución de Cualificacións para {modality}')
    plt.xlabel('Cualificación')
    plt.ylabel('Frecuencia')
    plt.xlim(0, 10)
    plt.ylim(0, 4) 
    plt.yticks(np.arange(0, 5, 1)) 

    plt.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()


plt.figure(figsize=(10, 6))
sns.boxplot(data=modalidad_df, x='Nom. Materia modalidade', y='Materia Modalidade', hue='Nom. Materia modalidade', palette=color_map, order=modalities, legend=False) # Use color map and specify order
sns.swarmplot(data=modalidad_df, x='Nom. Materia modalidade', y='Materia Modalidade', color='black', size=5, ax=plt.gca(), order=modalities) # Add individual points and specify order


modalities_mean = grouped_modalidad_df.mean().reindex(modalities) 
plt.scatter(x=range(len(modalities)), y=modalities_mean.values, color='white', edgecolors='black', s=100, zorder=3) 

plt.title('Box Plot de Cualificacións por Materia de Modalidade con Puntos Individuais e Medias')
plt.xlabel('Materia de Modalidade')
plt.ylabel('Cualificación')
plt.ylim(0, 10)
plt.yticks(np.arange(0, 11, 1)) 
plt.xticks(rotation=45, ha='right')


plt.grid(axis='y', linestyle='--')



plt.tight_layout()
plt.show()
```
