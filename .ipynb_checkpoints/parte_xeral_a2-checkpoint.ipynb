# A2. Proba xeral: diferenciando materias de modalidad

## 1. Identificar las materias de modalidad

Filtramos el DataFrame para incluir solo las columnas relevantes para las materias de modalidad, las cuales son: ´Materia de modalidade´ y ´Cualif. Materia de Modalidade´


```python
import pandas as pd

df = pd.read_excel("proba_xeral.xlsx")

modalidad_df = df[['Apelidos','Nome', 'Nom. Materia modalidade', 'Materia Modalidade']].copy()

display(modalidad_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the counts for each modality
modalidad_counts = modalidad_df['Nom. Materia modalidade'].value_counts()

plt.figure(figsize=(8, 5))
ax = sns.barplot(x=modalidad_counts.index, y=modalidad_counts.values, hue=modalidad_counts.index, palette='Set3')
plt.title('Número de Estudiantes con Calificación por Materia de Modalidad')
# plt.xlabel('Modality Subject')
plt.ylabel('Number of Students')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add numerical values on top of each bar
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python

modalidad_df.groupby('Nom. Materia modalidade')['Materia Modalidade'].describe().transpose()
```

## 2. Agrupar por materia de modalidad


Agrupar los datos por la columna 'Materia modalidade' para preparar el análisis estadístico y la visualización por materia.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Estadísticas descriptivas
modalidad_stats = modalidad_df['Materia Modalidade'].describe().to_frame().T
display(modalidad_stats)

# Ajustar estilo global
sns.set(style="whitegrid")

# Crear figura con 1 fila y 2 columnas
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # ancho ajustado a 10, altura a 4

# Histograma (izquierda)
sns.histplot(modalidad_df['Materia Modalidade'], bins=10, kde=True, color='lightcoral', ax=axes[0])
axes[1].set_title('Histograma de Calificaciones')
axes[1].set_xlabel('Calificación')
axes[1].set_ylabel('Frecuencia')

# Boxplot (derecha)
sns.boxplot(y=modalidad_df['Materia Modalidade'], color='lightcoral', ax=axes[1])
axes[0].set_title('Distribución (Boxplot)')
axes[0].set_ylabel('Calificación')
axes[0].set_ylim(0, 10)

plt.tight_layout()
plt.show()
```

## 2. Visualización por materia

Generar histogramas y box plots de la columna 'Materia Modalidade' para cada materia de modalidad para visualizar sus distribuciones.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

# Get unique modalities, sort alphabetically, and create a color mapping using 'Set3'
modalities = sorted(modalidad_df['Nom. Materia modalidade'].unique())
colors = sns.color_palette('Set3', len(modalities))
color_map = dict(zip(modalities, colors))

# Group the DataFrame by 'Nom. Materia modalidade'
grouped_modalidad_df = modalidad_df.groupby('Nom. Materia modalidade')['Materia Modalidade']

# Histograms
num_modalidades = len(modalities)
cols = 3
rows = math.ceil(num_modalidades / cols)

plt.figure(figsize=(cols * 5, rows * 4))

for i, modality in enumerate(modalities):
    plt.subplot(rows, cols, i + 1)
    group = grouped_modalidad_df.get_group(modality)
    # Create bins from 0 to 10 with a step of 1
    bins = np.arange(0, 11, 1)  # Bins from 0 to 10
    plt.hist(group, bins=bins, color=color_map[modality], edgecolor='black')  # Use color map
    plt.title(f'Distribución de Cualificacións para {modality}')
    plt.xlabel('Cualificación')
    plt.ylabel('Frecuencia')
    plt.xlim(0, 10)
    plt.ylim(0, 4)  # Set fixed y-axis limit
    plt.yticks(np.arange(0, 5, 1))  # Set fixed y-ticks

    plt.grid(axis='y', linestyle='--')

plt.tight_layout()
plt.show()

# Box plot with individual points and means
plt.figure(figsize=(10, 6))
sns.boxplot(data=modalidad_df, y='Nom. Materia modalidade', x='Materia Modalidade', hue='Nom. Materia modalidade', palette=color_map, order=modalities, legend=False)  # Use color map and specify order
sns.swarmplot(data=modalidad_df, y='Nom. Materia modalidade', x='Materia Modalidade', color='black', size=5, ax=plt.gca(), order=modalities)  # Add individual points and specify order

# Add mean points for each modality
modalities_mean = grouped_modalidad_df.mean().reindex(modalities)  # Reindex means to match sorted order
plt.scatter(y=range(len(modalities)), x=modalities_mean.values, color='white', edgecolors='black', s=100, zorder=3)  # Add white points with black edge

plt.title('Box Plot de Cualificacións por Materia de Modalidade con Puntos Individuais e Medias')
plt.ylabel('Materia de Modalidade')
plt.xlabel('Cualificación')
plt.xlim(0, 10)
plt.xticks(np.arange(0, 11, 1))  # Set x-ticks from 0 to 10 with step 1
plt.yticks(rotation=45, ha='right')

# Add horizontal grid lines
plt.grid(axis='x', linestyle='--')

plt.tight_layout()
plt.show()
```
